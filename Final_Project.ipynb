```python
from functools import total_ordering  
from math import log2  


@total_ordering  
class Char:
    def __init__(self, name, freq) -> None:  
        self._name = name
        self._freq = freq
        self._code = ""

    def __lt__(self, other):  
        return True if self._freq < other.get_freq()else False

    def __eq__(self, other): 
        return True if self._name == other.get_name() and self._freq == other.get_freq()else False

    def __str__(self):  
        
        if str(self._name) == ' ':
            return "{0}\t {1}\t {2}".format("space", '{0:.4f}'.format(float(self._freq)), self._code)
        return "{0}    \t {1}\t {2}".format(self._name, '{0:.4f}'.format(float(self._freq)), self._code)

    def __iter__(self):  
        return self

    def get_name(self):  
        return self._name

    def get_freq(self):  
        return self._freq

    def get_code(self):  
        return self._code

    def append_code(self, code): 
        self._code += str(code)


def find_middle(lst):
    if len(lst) == 1:
        return None
    s = k = b = 0
    for p in lst:
        s += p.get_freq()
    s /= 2
    for p in range(len(lst)):
        k += lst[p].get_freq()
        if k == s:
            return p
        elif k > s:
            j = len(lst) - 1
            while b < s:
                b += lst[j].get_freq()
                j -= 1
            return p if abs(s - k) < abs(s - b) else j
    return


def shannon_fano(lst):
    middle = find_middle(lst)
    if middle is None:
        return
    for i in lst[: middle + 1]:
        i.append_code(0)
    shannon_fano(lst[: middle + 1])
    for i in lst[middle + 1:]:
        i.append_code(1)
    shannon_fano(lst[middle + 1:])


def processing(text):
    lst = []  
    m = text  
    s = frozenset(m)  
    for c in s:
        lst.append(Char(c, m.count(c)/len(m)))

    lst.sort(reverse=True)  
    shannon_fano(lst)
    h = 0
    l = 0
    return lst
```

```python
def Shannon(our_dict):
    global df
    min1 = list(our_dict)[0]
    min2 = list(our_dict)[1]
    
    df[f'node{int((len(df) - len(our_dict))/2)}'] = [our_dict[min1] + our_dict[min2],
                                                            min1, min2, '']
    
    our_dict[f"node{int(((len(df) - 1) - len(our_dict))/2)}"] = our_dict[min1] + our_dict[min2] 
    
    del our_dict[min1]
    del our_dict[min2]
    if(len(our_dict)==1):
        return (our_dict, df) 
    else:
        return Shannon(our_dict)

agash = Shannon(our_dict)[1]

agash = dict(sorted(agash.items(), key=lambda a: a[1], reverse = True))

for key in agash:
    if agash[key][1] != 'None':
        agash[agash[key][1]][-1] = agash[key][-1] + '0'
        agash[agash[key][2]][-1] = agash[key][-1] + '1'
    else:
        pass
```

```python
def is_power_of_two(n): #degree of 2 : T or F
    return (n != 0) and (n & (n-1) == 0)
```

```python
def is_power_of_two(n):
    return (n != 0) and (n & (n-1) == 0)
```

```python
# Step 3
def HammingDecode74(bitstr2, length_of_block2):
    
    if len(bitstr2)!=length_of_block2:
        last_res2 = ''
        for i in range(1, len(bitstr2)):
            if is_power_of_two(i) == False:
                last_res2 = last_res2 + bitstr2[i]
        print(f'Decoded bitstring: {last_res2}')
        return last_res2
    
    # Get data bits
    n = 1
    b = [bitstr2[i:i+n]
              for i in range(0, len(bitstr2), n)]
    
    #to integer
    for i in range(0, len(b)):
        b[i] = int(b[i])
        
    check2 = ''
    check_res2 = ''
    
    #p1
    p1 = b[3]^b[5]^b[7]
    if p1 == b[1]:
        check_res2 = 'correct'
        sl = 'a'
    else:
        check_res2 = 'incorrect'
        sl = 'b'
    check2 += sl
    print(f'p1: b3+b5+b7= {b[3]}+{b[5]}+{b[7]} = {p1} {check_res2}')
    
    #p2
    p2 = b[3]^b[6]^b[7]
    if p2 == b[2]:
        check_res2 = 'correct'
        sl = 'a'
    else:
        check_res2 = 'incorrect'
        sl = 'b'
    check2 += sl
    print(f'p2: b3+b6+b7 = {b[3]}+{b[6]}+{b[7]} = {p2}  {check_res2}')
    
    #p4
    p4 = b[5]^b[6]^b[7]
    if p4 == b[4]:
        check_res2 = 'correct'
        sl = 'a'
    else:
        check_res2 = 'incorrect'
        sl = 'b'
    check2 += sl
    print(f'p4: b5+b6+b7 = {b[5]}+{b[6]}+{b[7]} = {p4} {check_res2}')
    
    #p0
    p0 = b[1]^b[2]^b[3]^b[4]^b[5]^b[6]^b[7]
    if p0 == b[0]:
        check_res2 = 'correct'
        sl = 'a'
    else:
        check_res2 = 'incorrect'
        sl = 'b'
    print(f'p0: b1+b2+b3+b4+b5+b6+b7 = {b[1]}+{b[2]}+{b[3]}+{b[4]}+{b[5]}+{b[6]}+{b[7]} = {p0} {check_res2}')
    ful = check2 + sl
    p_id_2 = -1
    if ful == 'aaaa':
        print('No error')
    else:
        if check2 == 'bbb': p_id_2 = 7
        elif check2 == 'abb': p_id_2 = 6
        elif check2 == 'bab': p_id_2 = 5
        elif check2 == 'aab': p_id_2 = 4
        elif check2 == 'bba': p_id_2 = 3
        elif check2 == 'aba': p_id_2 = 2
        elif check2 == 'baa': p_id_2 = 1
        elif check2 == 'aaa': p_id_2 = 0
    print(p_id_2)
    
    if ful != 'aaaa' and p_id_2 != -1:
        print(f'Error in position {p_id_2}')
        # correcting error
        b[p_id_2] = b[p_id_2]^1
        corr_bit = ''.join(map(str, b))
        print(f'Corrected bitstring: {corr_bit}')
    origina = [b[3], b[5], b[6], b[7]]
    decoded_str4 = ''.join(map(str, origina))
    print(f'Decoded bitstring: {decoded_str4}')
    return decoded_str4
  
```

```python
# Step 4
decoded_seq74 = ''
for i in range(0, len(data_block4)):
    print(str(data_block4[i]) + ':')
    dec74 = HammingDecode74(data_block4[i], len(data_block4[0]))
    decoded_seq74 = decoded_seq74 + str(dec74)
    print('----------')
```
